```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from fitds.fit2df import parse
```

```python
df = parse("../data/2020-01-24-20-05-57.fit")
```

```python
df.head()
```

```python
df["altitude"].plot()
```

```python
fig=plt.figure(figsize=(18, 10))
plt.plot(df["altitude"].values, label="Day 4")
plt.title("Elevation")
plt.style.use('seaborn-darkgrid')
plt.xlabel("Time")
plt.ylabel("Feet")
palette = plt.get_cmap('Set1')
plt.legend()
plt.show()
```

```python
elevation = df["altitude"].values
min_ylim = min(elevation) - 100
max_ylim = max(elevation) + 100
t_values = [x for x in range(len(df["altitude"]))]
```

```python
fig=plt.figure(figsize=(18, 10))
plt.plot(elevation, label="2020-01-24", color="darkgray")
plt.fill_between(t_values, min_ylim, elevation, color="lightgray")
plt.ylim(min_ylim, max_ylim)
plt.title("Climb")
plt.style.use('seaborn-darkgrid')
plt.xlabel("Elapsed Time (Seconds)")
plt.ylabel("Elevation (Feet)")
palette = plt.get_cmap('Set1')
plt.legend()
plt.show()
```
